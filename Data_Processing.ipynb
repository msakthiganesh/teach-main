### Copying images_and_states to data3

```python
images_source_dir = os.path.abspath("/Users/sakthi/Desktop/teach-main/data/images_and_states/images/valid_seen/")
images_destination_dir = os.path.abspath("/Users/sakthi/Desktop/teach-main/data4/images/valid_seen")
```

```python
images_f = os.listdir(images_source_dir)
```

```python
images_to_copy = []
for i, v in enumerate(file_names):
    for f in images_f:
        if v in f:
            images_to_copy.append(f)
```

```python
len(images_to_copy)
```

```python
for f in images_to_copy:   
    os.mkdir(os.path.join(images_destination_dir, f))
```

```python
for f in images_to_copy:
    f_lst = os.listdir(os.path.join(images_source_dir, f))
    src = images_source_dir +"/" + f
    dest = images_destination_dir +"/" + f
    for img in f_lst:
        shutil.copy(os.path.join(src, img), dest)
```
